```python
# Quiero un programa que genere un numero aleatorio del 1 al 100, 
# le pida al usuario que introduzca el numero y le diga si el numero es mas alto o mas bajo.
# Numero de intentos son 10. Si el numero que introduzca el usuario acierta antes del los intentos gana y muestra:
# "Has acertado, el numero es xx"


import random

num=random.randint(1,100)

numero = int(input("Introduzca un numero del 1 al 100: "))
print("Su numero introducido es: ", numero)

for i in range (5):
    if num == numero:
        print("Has acertado, el numero es:", num)
        print("Su numero introducido es: ", numero)
        break
    elif num > numero:
        print("El numero introducido es menor")
        numero = int(input("Introduzca otro numero: "))
        print("Su numero introducido es: ", numero)
    elif num < numero:
        print("El numero introducido es mayor")
        numero = int(input("Introduzca otro numero: "))
        print("Su numero introducido es: ", numero)
 
else:
    print("Has agotado los 5 intentos, suerte la próxima vez!")
        
```

```python
import random

num=random.randint(1,100)

intentos = 9

numero = int(input("Introduzca un numero del 1 al 100: "))

while intentos != 0:
    if num == numero:
        print("Has acertado, el numero es:", num)
        break
    elif num < numero:
        print("El numero introducido es mayor")
        numero = int(input("Introduzca otro numero: "))
        intentos -=1
        print("Te quedan", intentos, "intentos")
    elif num > numero:
        print("El numero introducido es menor")
        numero = int(input("Introduzca otro numero: "))
        intentos -=1
        print("Te quedan", intentos, "intentos")

else:
    print("Has perdido. El número era:", num)
```
